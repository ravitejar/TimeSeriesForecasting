```python
#Mean Forecast Error
def mean_forecast_err(y, yhat):
    return y.sub(yhat).mean()
```

```python
#Mean Absolute Errow
def mean_absolute_err(y, yhat):
    return np.mean((np.abs(y.sub(yhat).mean()) / yhat)) # or percent error = * 100
```

```python
print "MFE = ", mean_forecast_err(sales_1['Weekly_Sales'], predict_sunspots20)
print "MAE = ", mean_absolute_err(sales_1['Weekly_Sales'], predict_sunspots20)
```
